```python
import nltk
nltk.download('punkt')
```

```python
string0 = "The NLTK module is the natural language toolkit module."
```

```python
from nltk.tokenize import word_tokenize
```

```python
words0 = word_tokenize(string0)
words0
```

```python
length0 = len(words0)
length0
```

```python
string1 = "We can use it to tokenize strings into words or sentences."
words1 = word_tokenize(string1)
length1 = len(words1)
length1
```

```python
string2 = "We can then use the len() function to determine the number of words or sentences in a string."
words2 = word_tokenize(string2)
length2 = len(words2)
length2
# words2
```

```python
string3 = "So, again, we first tokenize the string into words or sentences. Then use the len() function to find the number of words or sentences in the string."
words3 = word_tokenize(string3)
length3 = len(words3)
length3
```

```python
string = """The NLTK module is the natural language toolkit module.
We can use it to tokenize strings into words or sentences.
We can then use the len() function to determine the number of words or sentences in a string.
So, again, we first tokenize the string into words or sentences.
Then use the len() function to find the number of words or sentences in the string."""

words = word_tokenize(string)
word_len = len(words)
word_len
```

```python
from nltk.tokenize import sent_tokenize
```

```python
sentences = sent_tokenize(string)
sent_len = len(sentences)
# sent_len
sentences
```
