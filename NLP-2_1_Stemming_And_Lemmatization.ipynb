# Stemming And Lemmatization



* **"Stemming and Lemmatization"** are text normalization techniques in **"Natural Language Processing (NLP)"**
* The main purposes of these *('Stemming and Lemmatization')* techniques are to prepare the words, texts and documents in a more simplified and standarized way, which helps in processing and enhances the efficiency of various NLP applications.
* Since we have various speaking languages used across the globe, these languages use tons of different words in multiple tones, pitches, and slang.
* Further, there are again tons of derived words from these basic language words, and these words are called as **"inflected language"**
* In order to bring these different language words into their basic root form, NLP is going to use *'Stemming and Lemmatization'*
* However, **lemma** is a root form of an actual language word, whereas **stem** may not be an actual language word.

```python
from stemming.porter2 import stem

documents = ["Human machine interface for lab abc computer applications",
             "A survey of user opinion of computer system response time",
             "The EPS user interface management system",
             "System and human system engineering testing of EPS",
             "Relation of user perceived response time to error measurement",
             "The generation of random binary unordered trees",
             "The intersection graph of paths in trees",
             "Graph minors IV Widths of trees and well quasi ordering",
             "Graph minors A survey"]

stem_words = [stem(word) for word in documents]
stem_words
```

```python
stem_words = [[stem(word)] for word in documents]
stem_words
```

```python
stem_words = [[stem(word) for word in sentence.split(" ")] for sentence in documents]
stem_words
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('all')
wnl = WordNetLemmatizer()
```

```python
# single word lemmatization examples
list1 = ['kites', 'babies', 'dogs', 'flying', 'smiling',
         'driving', 'died', 'tried', 'feet']
lemm_words = [wnl.lemmatize(word) for word in list1]
lemm_words
```

```python
lemm_words = [[wnl.lemmatize(word)] for word in list1]
lemm_words
```

```python
#print("{0}, {1}".format(lemm_words))

for word in list1:
  #print("{0} => {1}".format(word, wnl.lemmatize(word)))
  print(word, " ".ljust(15,"."),wnl.lemmatize(word))
```

```python
(print("{0} => {1}".format(word, wnl.lemmatize(word))) for word in list1)
```

```python
[print("{0} => {1}".format(word, wnl.lemmatize(word))) for word in list1]
```
