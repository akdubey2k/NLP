```python
phone_numbers = [phone0, phone1, phone2, phone3, phone4, phone5]
phone_numbers
```

```python
email_ids = [email_id0, email_id1, email_id2, email_id3, email_id4, email_id5, email_id6, email_id7, email_id8, email_id9, email_id10]
email_ids
```

```python
'''
\d{5}\s\d{5} :=>

\d{5} : any digit 5 times
\s    : whitespace character


\(\d{3}\)\s\d{3}\s\d{4} :=>
\(    : open bracket character "(" along with single escape character
\)    : close bracket character ")" along with single escape character
'''
ph_no_pattern = "\d{5}\s\d{5}|\d{10}|\(\d{3}\)\s\d{3}\s\d{4}"
# pattern0 = "\d{5}\s\d{5}|\d{10}|\(\d{3}\)\s\d{3}\s\d{4}|[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{3}|[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{2}\.[a-z]{2}|.+{5,8}"
# pattern0 = "\d{5}\s\d{5}|\d{10}|\(\d{3}\)\s\d{3}\s\d{4}|[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{3}|[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{2}\.[a-z]{2}|.+{5,8}"
```

```python
for i in range(len(phone_numbers)):
  if(match:=re.findall(ph_no_pattern, phone_numbers[i])):
    print(i, " : ", match)
```

```python
ph_no_match = [print(i, " : ", match) for i, number in enumerate(phone_numbers)\
               if (match:=re.findall(ph_no_pattern, number))]
# ph_no_match = [print(match) for i, number in enumerate(phone_numbers) if (match:=re.findall(pattern0, number))]
```

```python
'''
[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{3}  :=>
[a-zA-Z]* : zero or more character in the range of a-z or A-Z
\@        : special character "@" along with single escape character
\.        : special character "." along with single escape character

[a-zA-Z]{8,18}\@[a-z]{2,12}\.com :=>
[a-zA-Z]{8,18} : between 8 to 18 characters in the range of a-z or A-Z
'''
# email_id_pattern = "[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{3}|[a-zA-Z]*\@[a-zA-Z]*\.[a-z]{2}\.[a-z]{2}"
# email_id_pattern = "[a-zA-Z]{8,18}\@[a-z]{2,12}\.com|[a-zA-Z]{8,18}\@[a-z]{2,12}\.co.in"
email_id_pattern = "[a-zA-Z0-9_.-]{8,25}\@[a-z-]{2,12}.com|[a-zA-Z0-9_.-]{8,25}\@[a-z-]{2,12}.co.in"
```

```python
email_id_match = [print(i, " : ", match) for i, number in enumerate(email_ids) \
                  if(match:=re.findall(email_id_pattern, number))]
```
