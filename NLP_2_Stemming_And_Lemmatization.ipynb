## 1.1 Definition
## Stemming and Lemmatization
**Stemming** and **lemmatization** are both *text normalization techniques* used in Natural Language Processing (**NLP**) to reduce words to their base or root form. The purpose of these processes is to treat different forms of a word as the same word, which can be particularly useful in tasks like
* text mining,
* search engines, and
* information retrieval.


## 1.2 Stemming
**Stemming** is a crude (*not carefully made)* heuristic (*commonsense rule or general guess information*) process that chops off the ends of words in the hope of achieving the base form of the word called a **stem**.

**Stemming** usually involves removing common suffixes like **"ing", "ly", "s", "es", "ed",** and so on, which might not be a valid word.

### Example:
* "Playing" -> "Play"
* "Played" -> "Play"
* "Plays" -> "Play"
* "Runner" -> "Run"
* "Running" -> "Run"


## 1.3 Lemmatization
**Lemmatization** is a more advanced technique than **stemming**, as it takes into account the morphological (*related with the formation of admissible words in a language*) analysis of words to find the base or dictionary form, called a **lemma**.

**Lemmatization** considers the context of the word and its part of speech to determine the correct base form. This results in a more accurate and meaningful representation of the word.

### Example:
* "Running" -> "Run"
* "Better" -> "Good"
* "Was" -> "Be"
* "Feet" -> "Foot"

# 4. Import and initialized the PorterStemmer
* **stem,** this submodule of **nltk** includes various stemming algorithms.
* **PorterStemmer,** a stemming algorithms, which reduces words to their root form by removing or altering their suffixes.

```python
# A specific stemming algorithm from 'nltk' that reduces words to their base or root form.
from nltk.stem import PorterStemmer

# Creating an instance of PorterStemmer
ps = PorterStemmer()

# a list of words to stem using the Porter Stemmer.
stem_list = ['history', 'historical', 'finally', 'final', 'finalized', 'going', 'goes', 'gone']
```

## 4.1 Iterate the l

```python
# print(ps.stem(stem_list))
# ps.stem(stem_list) can take only 2 to 3 positional arguments at a time, so need to give in loop one-by-one
for word in stem_list:
  print(ps.stem(word))
```

```python
# print(ps.stem(word) for word in stem_list)
# list comprehension, has to go in list only
print([ps.stem(word) for word in stem_list])
```

```python
stem_words = ([ps.stem(word) for word in stem_list])
stem_words
```

#### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
lemm_list = ['kites', 'babies', 'dogs', 'flying', 'smiling', 'driving', 'died', 'tried', 'feet']
```

```python
for word in lemm_list:
  print(wnl.lemmatize(word))
```

```python
# list comprehension, has to go in list only
print([wnl.lemmatize(word) for word in lemm_list])
```

```python
lemm_words = [wnl.lemmatize(word) for word in lemm_list]
lemm_words
```
